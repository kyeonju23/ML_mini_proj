```python
import numpy as np
import pandas as pd

df_train = pd.read_csv("csv/train.csv")
df_train
```

```python
df_train.info()
```

```python
df_shuffled=df_train.iloc[np.random.permutation(df_train.index)].reset_index(drop=True)
```

```python
y_shuffled=df_shuffled["Activity"]
x_shuffled=df_shuffled.drop(["Activity"], axis=1)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation", scores.std())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression(solver="lbfgs", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
svm_clf = SVC(gamma="scale", random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')
```

```python
voting_clf.fit(x_shuffled, y_shuffled)
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier

extra_trees_clf = ExtraTreesClassifier(n_estimators=50, random_state=42)
extra_trees_clf.fit(x_shuffled, y_shuffled)
```

```python
from sklearn.model_selection import cross_val_score

extra_fold_score = cross_val_score(extra_trees_clf, x_shuffled, y_shuffled, cv=10, scoring="accuracy")
extra_fold_score
```

```python
display_scores(extra_fold_score)
```
